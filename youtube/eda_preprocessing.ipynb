# Exploratory Data Analysis (EDA) and Data Processing

## What is Exploratory Data Analysis (EDA)?

Exploratory Data Analysis (EDA) is the process of examining data sets to understand their characteristics, trends, and patterns.

## Key Steps in EDA

1. **Data Cleaning**: Remove missing or duplicate values, handle outliers, and perform data normalization.
2. **Data Transformation**: Convert data into a suitable format for analysis, such as aggregation or grouping.
3. **Data Visualization**: Use plots, charts, and heatmaps to visualize the data and identify trends or patterns.
4. **Summary Statistics**: Calculate summary statistics, such as mean, median, mode, and standard deviation.
5. **Correlation Analysis**: Examine the relationships between variables using correlation matrices.

## Benefits of EDA

1. **Identify Patterns and Trends**: EDA helps to uncover patterns and trends in the data that may not be immediately apparent.
2. **Improve Data Quality**: EDA can identify issues with data quality, such as outliers or missing values.
3. **Reduce the Complexity of the Data**: EDA simplifies the data by identifying irrelevant or redundant variables.
4. **Streamline the Analysis Process**: EDA provides a clear direction for further analysis.

## Common Data Processing Techniques

1. **Handling Missing Values**: Replace missing values with mean, median, or mode, or use advanced techniques like imputation.
2. **Data Normalization**: Scale data to a common range to prevent features with large ranges from dominating the analysis.
3. **Feature Selection**: Select a subset of features that are most relevant to the analysis.
4. **Data Coding**: Translate categorical data into numerical data for analysis.
5. **Handling Outliers**: Identify and remove or transform outliers that may skew the analysis.

## Why is EDA Important?

1. **Prevents Analysis Overfitting**: EDA helps to identify overfitting by selecting the most relevant variables for analysis.
2. **Improves Analysis Efficiency**: EDA streamlines the analysis process by identifying the most important relationships and patterns.
3. **Enables Better Decision Making**: EDA provides a deeper understanding of the data, enabling better decision making.

## Data Analysis Workflow

1. **Exploratory Data Analysis (EDA)**: Examine the data to understand its characteristics, trends, and patterns.
2. **Data Preprocessing**: Clean, transform, and prepare the data for analysis.
3. **Model Selection**: Select a suitable model for the problem at hand.
4. **Model Training**: Train the model on the prepared data.
5. **Model Evaluation**: Evaluate the performance of the model.

# Data Preprocessing.

```python
df.isnull().sum()
```

```python
#simple imputer mean, median, mode etc
# Iterative imputer - KNN, Decision tree, Random forest etc
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
numerical_imputer=Pipeline(steps=[('imputer',SimpleImputer(strategy='mean'))])
```

```python
scaler=Pipeline(steps=[('scaler',StandardScaler())])
```

```python
from sklearn.compose import ColumnTransformer
preprocessor=ColumnTransformer(transformers=[('num',numerical_imputer,X.columns),('scale',scaler,X.columns)],remainder='passthrough')
```

```python
pipe=Pipeline(steps=[('preprocessor',preprocessor)])
```

```python
pipe.fit(X_train)
```

```python
import pickle
pickle.dump(pipe,open('pipe.pkl','wb'))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
```

```python
X_train_transformed=pipe.transform(X_train)
X_test_transformed=pipe.transform(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix,roc_auc_score,roc_curve,precision_score,recall_score,f1_score
from sklearn.model_selection import cross_val_score
```

```python
def model_fit(model):
    model.fit(X_train_transformed,y_train)
    train_score=model.score(X_train_transformed,y_train)
    test_score=model.score(X_test_transformed,y_test)
    print(f'Train score is {train_score}')
    print(f'Test score is {test_score}')
    print(f'Classification report is \n{classification_report(y_test,model.predict(X_test_transformed))}')
    print(f'Confusion matrix is \n{confusion_matrix(y_test,model.predict(X_test_transformed))}')
    print(f'Accuracy score is {accuracy_score(y_test,model.predict(X_test_transformed))}')
    print(f'Precision score is {precision_score(y_test,model.predict(X_test_transformed),average="weighted")}')
    print(f'Recall score is {recall_score(y_test,model.predict(X_test_transformed),average="weighted")}')
    print(f'F1 score is {f1_score(y_test,model.predict(X_test_transformed),average="weighted")}')
    print(f'cross_val_score is {cross_val_score(model,X_train_transformed,y_train,cv=5).mean()}')
    
    
    return train_score,test_score
```

```python
lr=LogisticRegression()
```

```python
model_fit(lr)
```

```python
rf=RandomForestClassifier()
```

```python
model_fit(rf)
```

```python
svc=SVC()
```

```python
model_fit(svc)
```

```python
model_fit(DecisionTreeClassifier())
```

```python
model_fit(KNeighborsClassifier())
```
